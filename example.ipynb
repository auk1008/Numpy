## Real World Data Example

```python
import numpy as np
import matplotlib.pyplot as plt

# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
   [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
   [2, 120000, 140000, 160000, 110000],  # Beijing Bites
   [3, 200000, 230000, 260000, 300000],  # Pizza Hub
   [4, 180000, 210000, 240000, 270000],  # Burger Point
   [5, 160000, 185000, 205000, 230000]   # Chai Point
])
```

```python
# print (sales_data[:, 1])

# print (sales_data[:, 1:])


# Total sales per year

print ("Total Sales of all restaurants | yearly report:", np.sum(sales_data[:, 1:], axis=0))
print ("Total sales of a single resutaurants over the years [ 2021 - 2024 ]:", np.sum(sales_data[:, 1:], axis=1))
```

```python
# Note: If slicing result is a 1D array, then axis should be zero.
# There is no axis=1 in 1D array

# Min sale
min_sales = np.min(sales_data[0, 1:], axis=0)  # for Paradise Biryani
print (min_sales)

# Max sale
max_sales = np.max(sales_data[2, 1:], axis=0) # for pizza hub
print(max_sales)


# Average sales of all restaurants 
avg_sales = np.average(sales_data[:, 1:], axis=1)
print(avg_sales)
```

```python
# PLOTTING a graph

years = [2021, 2022, 2023, 2024]
restaurant_names = ["Paradise Biryani", "Beijing Bites", "Pizza Hub", "Burger Point", "Chai Point"]

# for i in range(len(sales_data)):
#    plt.plot(years, sales_data[i, 1:], label=restaurant_names[i])


plt.plot(years, sales_data[1, 1:])

plt.xlabel("Year")
plt.ylabel("Sales (in Rupees)")
plt.title("Restaurant Sales (2021-2024)")
# plt.legend()
plt.show()
```
